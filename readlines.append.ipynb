```python
with open("istiklal.txt", "r", encoding = "utf-8") as f:
    ist = f.readlines()
counter = 0
with open("istiklal.txt", "w", encoding = "utf-8") as f:
    for i in ist:
        counter += 1
        if counter % 4 == 0:
            f.write(i + "\n")
        else:
            f.write(i)
with open("istiklal.txt", "r", encoding = "utf-8") as f:
    print(f.read())
```
